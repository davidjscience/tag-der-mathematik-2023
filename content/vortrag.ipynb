```python
import piplite
await piplite.install('ipywidgets')
%matplotlib inline
from ipywidgets import interactive
from sympy import symbols,cos, sin, pi,factorial,ln
from sympy.plotting import plot
from sympy.plotting.plot import MatplotlibBackend, Plot
import matplotlib.pyplot as plt
def get_sympy_subplots(plot:Plot):
    backend = MatplotlibBackend(plot)
    backend.process_series()
    return backend.plt


x = symbols('x')

def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p
```

## Was wir außerdem wissen
\begin{align*}
&f(x)=\cos(x),\quad &f'(x)=-\sin(x),\\
&f''(x)=-\cos(x),\quad &f'''(x)=\sin(x),\\
&f''''(x)=\cos(x),\quad&\dots
\end{align*}

\begin{align*}
&f(0)=1,\quad &f'(0)=0,\\
&f''(0)=-1,\quad &f'''(x)=0,\\
&f''''(0)=1,\quad&\dots
\end{align*}

```python
def p(a_0,a_1,a_2):
    x_0 = 0
    p = plot(a_2*(x-(x_0))**2 + a_1*(x-x_0) + a_0, xlim=(-2*pi,2*pi),ylim=(-1,1), show=False)
    plt = get_sympy_subplots(p)
    plt.plot([x_0], [a_0], "o")
    plt.title('p(x)={a_0:.2f} + {a_1:.2f}x + {a_2:.2f}x^2 \n'.format(a_0=a_0,a_1=a_1,a_2=a_2),
              fontdict={'fontsize': 15})
    plt.show()

def parabel():    

    interactive_plot = interactive(p,a_2=(-2.0,2.0,.25),a_1=(-1,1,0.25),a_0=(-1,1,0.25))
    output = interactive_plot.children[-1]
    output.layout.height = '500px'
    return interactive_plot
```

## Taylorpolynom

\begin{align*}
p_2(x) &= a_0 + a_1 (x-x_0) + a_2 (x-x_0)^2\\
     &= f(x_0) + f'(x_0) (x-x_0) + \tfrac {f''(x_0)}{2}(x-x_0)^2
\end{align*}

## Taylorpolynom vom Grad 3


\begin{align*}
p_3(x) &= a_0 + a_1 (x-x_0) + a_2 (x-x_0)^2 + a_3 (x-x_0)^3\\
     &= f(x_0) + f'(x_0) (x-x_0) + \tfrac {f''(x_0)}{2}(x-x_0)^2 + \tfrac {f''(x_0)}{6}(x-x_0)^3
\end{align*}

## Taylorpolynom vom Grad n

\begin{align*}
p_n(x) &= a_0 + a_1 (x-x_0) + a_2 (x-x_0)^2  + \dots + a_n (x-x_0)^n \\
     &= f(x_0) + f'(x_0) (x-x_0) + \tfrac {f''(x_0)}{2}(x-x_0)^2 + \dots + \tfrac {f^{(n)}(x_0)}{n!}(x-x_0)^n
\end{align*}

```python
def l(x_0,n):
    p = plot(cos(x), taylor(cos(x), x_0, n), xlim=(-3*pi,3*pi),ylim=(-1,1), show=False)
    plt = get_sympy_subplots(p)
    plt.plot([x_0], [cos(x_0)], "o")
    plt.show()
    

def cosinus_taylor_approximation():
    
    interactive_plot = interactive(l,x_0=(round(-pi/2,6),round(pi/2,6),round(pi/6,6)), n=(1,25,1) )
    output = interactive_plot.children[-1]
    output.layout.height = '500px'
    return interactive_plot

```

```python
def l(n):
    p = plot(ln(x), taylor(ln(x), 1, n), xlim=(-3,3),ylim=(-5,5), show=False)
    plt = get_sympy_subplots(p)
    plt.plot([1], [ln(1)], "o")
    plt.show()
    
def logarithmus_taylor_approximation():
    
    interactive_plot = interactive(l, n=(1,26,1) )
    output = interactive_plot.children[-1]
    output.layout.height = '500px'
    return interactive_plot
```
